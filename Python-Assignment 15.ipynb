```python
# ATM Machine Simulator

user = {
    "user1": {
        "pin": 1234,
        "balance": 10000
    }
}

user_id = input("Enter User ID: ")
pin = int(input("Enter PIN: "))

if user_id in user and user[user_id]["pin"] == pin:

    while True:
        print("\n1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            print("Balance:", user[user_id]["balance"])

        elif choice == "2":
            amount = float(input("Enter amount to deposit: "))
            user[user_id]["balance"] += amount
            print("Amount deposited successfully")

        elif choice == "3":
            amount = float(input("Enter amount to withdraw: "))

            if amount <= user[user_id]["balance"]:
                user[user_id]["balance"] -= amount
                print("Please collect your cash")
            else:
                print("Insufficient balance")

        elif choice == "4":
            print("Thank you")
            break

        else:
            print("Invalid choice")

else:
    print("Wrong User ID or PIN")
```

```python
students = {}

n = int(input("How many students? "))

for i in range(n):

    name = input("Enter student name: ")

    marks = []

    for j in range(3):
        mark = float(input(f"Enter mark {j+1}: "))
        marks.append(mark)

    students[name] = marks

highest_avg = 0
topper = ""

total_class_avg = 0

for name, marks in students.items():

    total = sum(marks)
    avg = total / len(marks)

    print("\nName:", name)
    print("Total:", total)
    print("Average:", avg)

    total_class_avg += avg

    if avg > highest_avg:
        highest_avg = avg
        topper = name

class_avg = total_class_avg / len(students)

print("\nClass Average:", class_avg)
print("Topper:", topper)
```

```python
items = {
    "Apple": 50,
    "Milk": 30,
    "Bread": 40
}

cart = []

while True:

    print("\nAvailable Items")

    for item, price in items.items():
        print(item, "-", price)

    choice = input("Add item (or type exit): ")

    if choice.lower() == "exit":
        break

    if choice in items:
        cart.append(choice)
        print("Added to cart")
    else:
        print("Item not found")

total = 0

print("\nItems in Cart")

for item in cart:
    print(item)
    total += items[item]

if total > 100:
    discount = total * 0.10
    total = total - discount

print("Final Bill:", total)
```

```python
contacts = {}

while True:

    print("\n1.Add")
    print("2.Update")
    print("3.Delete")
    print("4.Search")
    print("5.Display")
    print("6.Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        name = input("Name: ")
        phone = input("Phone: ")
        email = input("Email: ")

        contacts[name] = {
            "phone": phone,
            "email": email
        }

    elif choice == "2":

        name = input("Enter name to update: ")

        if name in contacts:
            contacts[name]["phone"] = input("New phone: ")
            contacts[name]["email"] = input("New email: ")

    elif choice == "3":

        name = input("Enter name to delete: ")

        if name in contacts:
            del contacts[name]

    elif choice == "4":

        name = input("Enter name: ")

        if name in contacts:
            print(contacts[name])
        else:
            print("Contact not found")

    elif choice == "5":

        for name in sorted(contacts):
            print(name, contacts[name])

    elif choice == "6":
        break

    else:
        print("Invalid choice")
```

```python
books = ["Python", "Java", "C++"]
issued_books = []

while True:

    print("\n1. Add Book")
    print("2. Issue Book")
    print("3. Return Book")
    print("4. Search Book")
    print("5. Display Books")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        book = input("Enter book name: ")
        books.append(book)

    elif choice == "2":

        book = input("Enter book to issue: ")

        if book in books:
            books.remove(book)
            issued_books.append(book)
            print("Book issued")

        else:
            print("Book not available")

    elif choice == "3":

        book = input("Enter book to return: ")

        if book in issued_books:
            issued_books.remove(book)
            books.append(book)

    elif choice == "4":

        book = input("Enter book name: ")

        if book in books:
            print("Book available")

        elif book in issued_books:
            print("Book already issued")

        else:
            print("Book not found")

    elif choice == "5":

        print("Available Books:", books)
        print("Issued Books:", issued_books)

    elif choice == "6":
        break

    else:
        print("Invalid choice")
```

```python
transactions = []
balance = 0

while True:

    print("\n1. Deposit")
    print("2. Withdraw")
    print("3. Transaction History")
    print("4. Check Balance")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        amount = float(input("Enter deposit amount: "))

        transactions.append({
            "type": "Deposit",
            "amount": amount
        })

        balance += amount

    elif choice == "2":

        amount = float(input("Enter withdrawal amount: "))

        if amount <= balance:

            transactions.append({
                "type": "Withdraw",
                "amount": amount
            })

            balance -= amount

        else:
            print("Insufficient balance")

    elif choice == "3":

        print("\nTransaction History")

        for t in transactions:
            print(t)

    elif choice == "4":

        print("Current Balance:", balance)

    elif choice == "5":
        break

    else:
        print("Invalid choice")
```

```python
inventory = {}

while True:

    print("\n1. Add Product")
    print("2. Update Quantity")
    print("3. Display Inventory")
    print("4. Total Inventory Value")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        name = input("Product Name: ")
        quantity = int(input("Quantity: "))
        price = float(input("Price: "))

        inventory[name] = {
            "quantity": quantity,
            "price": price
        }

    elif choice == "2":

        name = input("Enter product name: ")

        if name in inventory:
            new_qty = int(input("New Quantity: "))
            inventory[name]["quantity"] = new_qty

    elif choice == "3":

        for product, details in inventory.items():
            print(product, details)

    elif choice == "4":

        total = 0

        for product in inventory:
            total += (
                inventory[product]["quantity"] *
                inventory[product]["price"]
            )

        print("Total Inventory Value =", total)

    elif choice == "5":
        break

    else:
        print("Invalid choice")
```

```python
questions = [
    {
        "question": "What is the capital of India?",
        "answer": "Delhi"
    },
    {
        "question": "How many days are there in a week?",
        "answer": "7"
    },
    {
        "question": "Python is a programming language? (yes/no)",
        "answer": "yes"
    }
]

score = 0

for q in questions:

    print("\n" + q["question"])

    user_answer = input("Your Answer: ")

    if user_answer.lower() == q["answer"].lower():
        print("Correct")
        score += 1

    else:
        print("Wrong")

print("\nFinal Score =", score)
print("Out of", len(questions))
```

```python
tasks = []

while True:

    print("\n1. Add Task")
    print("2. Delete Task")
    print("3. Mark Completed")
    print("4. View Tasks")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":

        task = input("Enter task: ")

        tasks.append({
            "task": task,
            "done": False
        })

    elif choice == "2":

        task = input("Enter task to delete: ")

        for t in tasks:

            if t["task"] == task:
                tasks.remove(t)

    elif choice == "3":

        task = input("Enter completed task: ")

        for t in tasks:

            if t["task"] == task:
                t["done"] = True

    elif choice == "4":

        for t in tasks:

            status = "Done" if t["done"] else "Pending"

            print(t["task"], "-", status)

    elif choice == "5":
        break

    else:
        print("Invalid choice")
```

```python
candidates = {
    "Ravi": 0,
    "Kiran": 0,
    "Anil": 0
}

voted_users = []

while True:

    voter = input("\nEnter voter name (or exit): ")

    if voter.lower() == "exit":
        break

    if voter in voted_users:
        print("You have already voted")
        continue

    print("\nCandidates")

    for candidate in candidates:
        print(candidate)

    vote = input("Vote for: ")

    if vote in candidates:

        candidates[vote] += 1
        voted_users.append(voter)

        print("Vote recorded")

    else:
        print("Invalid candidate")

winner = max(candidates, key=candidates.get)

print("\nVoting Results")

for candidate, votes in candidates.items():
    print(candidate, ":", votes)

print("Winner is", winner)
```

```python
seats = [
    ["A1", "A2", "A3"],
    ["B1", "B2", "B3"]
]

booked = []

ticket_price = 150

while True:

    print("\nAvailable Seats")

    for row in seats:
        print(row)

    seat = input("Enter seat number to book (or exit): ")

    if seat.lower() == "exit":
        break

    found = False

    for row in seats:

        if seat in row:

            row[row.index(seat)] = "X"
            booked.append(seat)

            print("Seat booked successfully")
            found = True

    if not found:
        print("Seat not available")

print("\nBooked Seats:", booked)
print("Total Amount =", len(booked) * ticket_price)
```

```python
text = input("Enter a paragraph: ")

text = text.lower()

words = text.split()

unique_words = set(words)

print("Number of Unique Words =", len(unique_words))

frequency = {}

for word in words:

    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

sorted_words = sorted(
    frequency.items(),
    key=lambda x: x[1],
    reverse=True
)

print("\nTop 5 Frequent Words")

for word, count in sorted_words[:5]:
    print(word, ":", count)
```

```python
primes = []

for num in range(2, 101):

    is_prime = True

    for i in range(2, num):

        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        primes.append(num)

print(primes)
```
